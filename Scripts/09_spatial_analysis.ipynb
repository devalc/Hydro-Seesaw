```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import parquet
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Specify the path to the extracted Parquet file
parquet_file = "/Users/sidchaudhary/Documents/GitHub/Hydro-Seesaw/Results/merged_flags_and_grdc_with_riverATLAS.parquet"

# Read the Parquet file into a DataFrame
df = pd.read_parquet(parquet_file)
```

```python
df.head(10)
#df.shape
```

```python

# Define the number of clusters
num_clusters = 10

# List to store clustering results
results = []

df_clustered = df.copy()
```

```python
for scenario in df['scenario'].unique():
    for time_period in df['time_period'].unique():
        # Filter the DataFrame based on the combination
        filtered_df = df[(df['scenario'] == scenario) & (df['time_period'] == time_period)]
        
        # Ensure there are enough samples for clustering
        if len(filtered_df) > num_clusters:
            # Prepare the data for clustering
            X = filtered_df[['p99_flag']].values
            
            # Standardize the data (important for clustering)
            scaler = StandardScaler()
            X_scaled = scaler.fit_transform(X)
            
            # Initialize KMeans with the desired number of clusters
            kmeans = KMeans(n_clusters=num_clusters, random_state=0)
            
            # Fit the KMeans model and predict clusters
            filtered_df['cluster'] = kmeans.fit_predict(X_scaled)
            
            # Merge cluster labels back into the main DataFrame
            df_clustered.loc[filtered_df.index, 'cluster'] = filtered_df['cluster']
        else:
            print(f"Not enough data to cluster for combination: {scenario}, {time_period}")

# Print or inspect the resulting DataFrame
print(df_clustered.head())
```
